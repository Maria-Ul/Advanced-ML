```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
from keras.datasets import mnist
(X_train, y_train), (X_pred, y_pred) = mnist.load_data()
```

```python
X_train
```

```python
np.array(X_train).shape
```

```python
X_train_1 = np.reshape(X_train, (60000, 28*28))
```

```python
X_train_1
```

```python
pca = PCA()
pca.fit((X_train_1))
```

```python
pca.explained_variance_ratio_[0:10]
```

```python
round(sum(pca.explained_variance_ratio_[0:52]),3)
```

```python
pca_52 = PCA(n_components=52, svd_solver='full').fit(X_train_1)
```

```python
fi = (pca_52.components_)
```

```python
Z = np.matmul(X_train_1, fi.transpose())
```

```python
X = pca_52.fit_transform(X_train_1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_data, X_test_data, y_train_data, y_test_data = train_test_split(X, y_train, test_size=0.3, random_state=126)
```

```python
X_train_data
```

```python
X_train_data.shape
```

```python
y_train_data
```

```python
np.mean(X_train_data[:,0])
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = OneVsRestClassifier(RandomForestClassifier(criterion='gini', min_samples_leaf=10, max_depth=20,n_estimators=10, random_state=126)).fit(X_train_data, y_train_data)
```

```python
pred = clf.predict(X_test_data)
count = 0
for i in range(len(pred)):
  if y_test_data[i]==5:
    if pred[i]==5:
      count+=1
print(count)
```
